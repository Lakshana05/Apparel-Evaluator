```python
import pandas as pd
import json

# Function to extract sizes from the JSON-like format
def extract_sizes(json_str):
    try:
        sizes_list = json.loads(json_str)
        sizes = [entry['available_sizes'] for entry in sizes_list]
        return ', '.join(sizes)
    except (json.JSONDecodeError, KeyError):
        return None

# Apply the function to the 'available_sizes' column
men_clothings['extracted_sizes'] = men_clothings['available_sizes'].apply(extract_sizes)

# Display the result
print(men_clothings[['available_sizes', 'extracted_sizes']])
```

```python
import pandas as pd
import json

# Function to extract sizes from the JSON-like format
def extract_specs(json_str):
    try:
        sizes_list = json.loads(json_str)
        sizes = [entry['specifications'] for entry in sizes_list]
        return ', '.join(sizes)
    except (json.JSONDecodeError, KeyError):
        return None

# Apply the function to the 'available_sizes' column
men_clothings['extracted_specs'] = men_clothings['specifications'].apply(extract_specs)

# Display the result
print(men_clothings[['specifications', 'extracted_specs']])
```

```python
import pandas as pd
import json

# Function to extract sizes from the JSON-like format
def extract_product_details(json_str):
    try:
        sizes_list = json.loads(json_str)
        sizes = [entry['prod_details'] for entry in sizes_list]
        return ', '.join(sizes)
    except (json.JSONDecodeError, KeyError):
        return None

# Apply the function to the 'available_sizes' column
men_clothings['extracted_product_details'] = men_clothings['prod_details'].apply(extract_product_details)

# Display the result
print(men_clothings[['prod_details', 'extracted_product_details']])
```
