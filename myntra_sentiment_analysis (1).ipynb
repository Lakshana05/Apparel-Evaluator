```python
import pandas as pd
import emoji
import re



# Function to remove emojis from text
def remove_emojis(text):
    return emoji.demojize(text)

# Apply the function to the 'review' column and create a new 'review_clean' column
women['review_clean'] = women['reviews'].apply(remove_emojis)

# Display the updated DataFrame
print(women.head())
```

```python
from transformers import pipeline

# Specify the model name and revision
model_name = "juliensimon/reviews-sentiment-analysis"


# Create sentiment analysis pipeline
sentiment_analysis = pipeline("sentiment-analysis", model=model_name)

# Example usage
review_sentiment=[]
reviews=women['review_clean']
for review in reviews:
  result = sentiment_analysis(review)
  labels = result[0]['label']
  if (labels == 'LABEL_0'):
    review_sentiment.append('NEGATIVE')
  else:
    review_sentiment.append('POSITIVE')



print(review_sentiment)
```

```python
women['review_sentiment']=review_sentiment
```

```python
men=pd.read_csv(r'/content/drive/MyDrive/myntra_csv/men_final.csv')
women=pd.read_csv(r'/content/drive/MyDrive/myntra_csv/women_final.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time

# Sample dataset
df = men[['brand','review_sentiment']]

# Function to update and display the race bar plot dynamically
def race_bar_plot(df):
    plt.figure(figsize=(10, 6))

    # Filter positive reviews
    positive_reviews = df[df['review_sentiment'] == 'POSITIVE']

    # Plotting the bar chart
    plt.barh(positive_reviews['brand'], [1] * len(positive_reviews), color='green', label='Positive Reviews')

    plt.xlabel('Count of Positive Reviews')
    plt.title('Race Bar Plot for Positive Reviews by Brand')
    plt.legend()
    plt.show()

# Dynamic update of the race bar plot
for i in range(len(df) + 1):
    race_bar_plot(df.head(i))
    clear_output(wait=True)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset

df = men[['brand','review_sentiment']]

# Count the reviews by brand and sentiment
review_counts = df.groupby(['brand', 'review_sentiment']).size().unstack(fill_value=0)

# Plotting the bar plot
plt.figure(figsize=(10, 6))
review_counts.plot(kind='bar', stacked=True, color=['red', 'green'])

plt.xlabel('Brand')
plt.ylabel('Count')
plt.title('Count of Positive and Negative Reviews by Brand')
plt.legend(title='Reviews', loc='upper right')
plt.show()
```
